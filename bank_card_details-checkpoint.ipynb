```python
import pandas as pd
df = pd.read_csv('BankChurners.csv')
df.head()
```

```python
df.info()
df.describe()
df.columns
```

## Cleaning and Preprocessing

```python
df_cleaned = df.drop(columns=[col for col in df.columns if "Naive_Bayes" in col])
```

```python
df_cleaned.isnull().sum()
df_cleaned.duplicated().sum()
```

```python
df_cleaned['credit_utilization'] = df_cleaned['total_revolving_bal'] / df_cleaned['credit_limit']
```

```python
df_cleaned.rename(columns=lambda x: x.strip().replace(" ", "_").lower(), inplace=True)
```

```python
df_cleaned.rename(columns=lambda x: x.strip().replace(" ", "_").lower(),
                  inplace=True)
```

```python
df_cleaned["credit_utilization"] = (
    df_cleaned["total_revolving_bal"] / df_cleaned["credit_limit"]
)
```

```python
df_cleaned["credit_utilization"].describe()
```

```python
df_cleaned["high_util_flag"] = (df_cleaned["credit_utilization"] > 0.8).astype(int)
```

## EXPLORATORY DATA ANALYSIS (EDA) (Seaborn, Matplotlib)

```python
df_cleaned['attrition_flag'].value_counts(normalize=True).plot(kind='bar')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Churn vs. Utilization
sns.boxplot(x='attrition_flag', y='credit_utilization', data=df_cleaned)
plt.title('Credit Utilization by Customer Status')
plt.show()
```

```python
# Monthly Transactions by Category
sns.histplot(df_cleaned['total_trans_ct'], kde=True)
plt.title('Distribution of Monthly Transactions per Customer')
plt.xlabel('Total Transactions in 12 Months')
plt.show()
```

```python
sns.histplot(df_cleaned['credit_utilization'], bins=30, kde=True)
```

```python
sns.boxplot(data=df_cleaned, x='attrition_flag', y='months_inactive_12_mon')
```

## FEATURE ENGINEERING

```python
df_cleaned['risk_flag'] = (
    (df_cleaned['credit_utilization'] > 0.8) &
    (df_cleaned['months_inactive_12_mon'] > 4)
).astype(int)
```

```python
df_cleaned['risk_flag'].value_counts()
```

```python
df_cleaned.groupby('risk_flag').mean(numeric_only=True)
```

```python
sns.countplot(x='Attrition_Flag', hue='risk_flag', data=df_cleaned)
```
